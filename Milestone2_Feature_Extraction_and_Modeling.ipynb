```python
import pandas as pd

# Load the CSV file
data = pd.read_csv('/content/data.csv')

# Display first 5 rows
data.head()
```

```python
data.info()
```

```python
# Remove duplicate rows
data = data.drop_duplicates()

# Check missing values
data.isnull().sum()
```

```python
numeric_data = data.select_dtypes(include=['int64', 'float64'])
numeric_data.head()
```

```python
data[data['anomaly'] == -1].head()
```

```python
features = pd.DataFrame()

features['mean'] = numeric_data.mean(axis=1)
features['std'] = numeric_data.std(axis=1)
features['min'] = numeric_data.min(axis=1)
features['max'] = numeric_data.max(axis=1)
features['range'] = features['max'] - features['min']

features.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(contamination=0.05, random_state=42)
features['anomaly'] = model.fit_predict(scaled_features)
```

```python
features['anomaly'].value_counts().plot(kind='bar')
plt.title('Normal vs Anomalies')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(features['mean'], features['std'],
            c=features['anomaly'], cmap='coolwarm')
plt.xlabel('Mean')
plt.ylabel('Standard Deviation')
plt.title('Anomaly Visualization (Feature Space)')
plt.colorbar(label='Anomaly')
plt.show()
```

```python
features['mean'].hist(bins=30)
plt.title('Distribution of Mean Feature')
plt.xlabel('Mean')
plt.ylabel('Frequency')
plt.show()
```

```python
X = features[['mean', 'std', 'range']]
X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
features['cluster'] = kmeans.fit_predict(X_scaled)

features.head()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_features = pca.fit_transform(X_scaled)

features['pca1'] = pca_features[:, 0]
features['pca2'] = pca_features[:, 1]
```

```python
plt.figure(figsize=(8,6))
plt.scatter(features['pca1'], features['pca2'],
            c=features['cluster'], cmap='viridis')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('Behavioral Clustering of Fitness Data')
plt.colorbar(label='Cluster')
plt.show()
```

```python
features['cluster'].value_counts().plot(kind='bar')
plt.title('Cluster Distribution')
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.show()
```
